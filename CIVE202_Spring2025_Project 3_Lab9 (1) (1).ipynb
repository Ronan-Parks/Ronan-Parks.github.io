## Bar Plot  
### Variables: Vehicle Make and Vehicle Year

```python
nhts_data = nhts_data.dropna(subset = ['make'])


data_grouped = nhts_data.groupby(['make', 'vehicle_year']).size().unstack(fill_value = 0)

# Plotting

plt.figure(figsize = (20,10)) # Graph Size

# Plotting stacked bar chart

data_grouped.plot(kind = 'bar', stacked = True, color = sns.color_palette('viridis', len(data_grouped.columns)))

# Customizing the plot

plt.xlabel('Make', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.title('Stacked Bar Chart of Make', fontsize = 18, fontweight='bold')
plt.xticks(rotation = 90)
plt.legend(title = 'Vehicle Year', bbox_to_anchor = (1,1), fontsize = 8)
#Show the plot
```

## Histogram 
### Variables: Vehicle Age 

```python
plt.figure(figsize=(12,6)) #graph size 
sns.histplot(x='vehicle_age', data = nhts_data, bins=20, color='Red') # histplot graphs a histogram 
#graph value, data set it is pulling from, number of bins, graph color
plt.xlabel("Vehicle Age", fontweight='bold') # X Axis Title 
plt.ylabel("Quantity", fontweight='bold') # Y Axis Title 
plt.title("Vehicle Age Distribution", fontweight='bold', fontsize=18) #Graph Title 
plt.xticks(rotation=30)  # Rotate x-axis labels 30 degrees 
plt.show() 
```

## Time-Series Plots 
### Variables Plot One: Time, Follower Speed, and Leader Speed 
### Variables Plot Two: Time, Follower Speed, and Follower Position 

```python
trajectory_number = 1
data_subset = ngsim_data[ngsim_data['trajectory_number']==1]
```

```python
plt.figure(figsize=(12,6)) # Graph Size
sns.lineplot(x='Time', y='follower_speed(m/s)', data = data_subset) # lineplot graphs a line 
sns.lineplot(x='Time', y='leader_speed(m/s)', data = data_subset) # X Axis Value, Y Axis Value, and data set it is pulling from 
plt.xlabel("Time", fontweight='bold') # X Axis Title
plt.ylabel("Position", fontweight='bold') # Y Axis Title 
plt.title("Follower and Leader Speed vs Time", fontweight='bold', fontsize=18) #Graph Title 
plt.xticks(rotation=45)  # Rotate x-axis labels 45 degrees 
plt.show()
```

```python
plt.figure(figsize=(12,6)) # Graph Size
sns.scatterplot(x='Time', y='follower_position(m)', hue='follower_speed(m/s)', data = data_subset, palette='rocket') # scatterplot graphs a scatter plot 
# X Axis Value, Y Axis Value, Assigning a hue based off of follower speed, data set it is pulling from, and selecting a specific color group
plt.xlabel("Time", fontweight='bold') # X Axis Title
plt.ylabel("Position", fontweight='bold') # Y Axis Title 
plt.title("Follower Position vs Time with Speed Intensity", fontweight='bold', fontsize=18) #Graph Title 
plt.xticks(rotation=45)  # Rotate x-axis labels 45 degrees 
plt.show()
```

## Box Plot 
### Variables: Vehicle Year and Vehicle Type 

```python
plt.figure(figsize=(12,6)) # Graph Size 
sns.boxplot(x='vehicle_year', y='vehicle_type', data = nhts_data) # X Axis Value, Y Axis Value, and data set it is pulling from
plt.xlabel("Vehicle Year", fontweight='bold') # X Axis Title
plt.ylabel("Vehicle Type", fontweight='bold') # Y Axis Title 
plt.title("Vehicle Year vs Type", fontweight='bold', fontsize=18) #Graph Title 
plt.xticks(rotation=30)  # Rotate x-axis labels 30 degrees 
plt.show()
```
